```python
x=df.drop(columns=["Attrition"],axis=1)
x.head()
```

```python
x.shape
```

```python
type(x)
```

```python
y=df["Attrition"]
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x["BusinessTravel"]=le.fit_transform(x["BusinessTravel"])
```

```python
x["Department"]=le.fit_transform(x["Department"])
```

```python
x["EducationField"]=le.fit_transform(x["EducationField"])
```

```python
x["Gender"]=le.fit_transform(x["Gender"])
```

```python
x["JobRole"]=le.fit_transform(x["JobRole"])
```

```python
x["MaritalStatus"]=le.fit_transform(x["MaritalStatus"])
```

```python
x["OverTime"]=le.fit_transform(x["OverTime"])
```

```python
x.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
```

```python
x_scaled=pd.DataFrame(ms.fit_transform(x),columns=x.columns)
```

```python
x_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size =0.2,random_state =0)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

## Evaluation Metrics

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
pd.crosstab(y_test,y_pred)
```

```python
#accuracy
(194+10)/239
```

```python
print(classification_report(y_test,y_pred))
```

```python
probability=model.predict_proba(x_test)[:,1]
```

```python
probability
```

```python
y_test = y_test.map({'Yes': 1, 'No': 0}).astype(int)
```

```python
fpr,tpr,threshsholds = roc_curve(y_test,probability)
```

```python
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```
