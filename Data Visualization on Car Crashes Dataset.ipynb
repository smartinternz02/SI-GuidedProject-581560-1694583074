## Understanding the dataset

In the given car crashes dataset, we analyse car crashes in the USA. The data provided is of the cause of the crashes and the cost to the insurance company.

The column **total** gives the percent of drivers involved in fatal collisions.<br>
The column **speeding** gives the percent of fatal collisions caused by speeding.<br>
The column **alcohol** gives the percent of fatal collisions caused by consumption of alcohol.<br>
The column **not_distracted** gives the percent of fatal collisions where the driver was not distracted.<br>
The column **no_previous** gives the percent of drivers not invloved in any previous collisions.<br>
The column **ins_premium** gives the amount of car insurance premium paid.<br>
The column **ins_losses** gives the losses incured by the car insurance companies.<br>
The column **abbrev** gives the names of the respective USA states for which the data is collected.

## Visualizing the data

### Visualizing using Matplotlib

#### Line Graph

```python
plt.figure().set_figwidth(15)
plt.plot(df.abbrev,df.total,'-*')
plt.grid()
plt.show()
```

**Inference:**<br>The above graph shows that the total number of accidents are maximum in the states of *Arkansas, North Dakota, South Carolina and West Virginina.*<br> It also shows that the accident percentages are lower in the state of *Washington DC and Massachusetts*


#### Subplot

```python
x=["AR", "ND", "SC", "WV"]
a=df.loc[df['abbrev'].isin(x),["speeding"]]
b=df.loc[df['abbrev'].isin(x),["alcohol"]]
plt.figure().set_figwidth(15)
plt.subplot(1,2,1)
plt.plot(x,a,'-*')
plt.title("Accidents caused by speeding")
plt.grid()
plt.subplot(1,2,2)
plt.plot(x,b,'r-^')
plt.title("Accidents caused by alcohol")
plt.grid()
plt.show()
```

**Inference:**<br> The above subplots show the comparison between the percentage of accidents caused by speeding and alcohol, respectively, in the states where the total number of accidents is greater. <br> It can be inferred that South Carolina still has a large number of accidents caused by speeding and alcohol, whereas in North Dakota, the accidents caused by speeding are less in comparison to those caused by driving under the influence.

#### Scatter Plots

```python
plt.scatter(df.ins_premium,df.ins_losses,marker='x',c='k')
plt.xlabel('Premium Paid')
plt.ylabel('Losses Incured')
```

**Inference:**<br> From this scatter plot of the car insurance premium paid against the losses incured by the car insurance company, it can be understood that the amount of losses incured is directly proportional to the premium paid. <br> Subsequently, as the amount of premium paid increases, the graph becomes more sparse, indicating that hign premium is being paid in a smaller number of states. Similarly, for the losses suffered, high losses are rarely incurred, helping us conclude that accidents with greater damage are rarer than those with less damage.

#### Compare Plots

```python
plt.figure().set_figwidth(15)
x=["CA","CT","DC","MA","MN","NH","NJ","NY","RI","UT","WA"]
y=df.loc[df['abbrev'].isin(x),["no_previous"]]
plt.scatter(x,y,c='teal',s=70)
y=df.loc[df['abbrev'].isin(x),["not_distracted"]]
plt.scatter(x,y,c='palegreen')
y=df.loc[df['abbrev'].isin(x),["speeding"]]
plt.scatter(x,y,c='coral')
y=df.loc[df['abbrev'].isin(x),["alcohol"]]
plt.scatter(x,y,c='mediumvioletred')
```

**Inference:**<br> The graph above is a comparison of the percent of drivers involved in accidents and the characteristics of the accident and the driver. <br> 7 out of 11 states show that more accidents are caused by speeding than by DUI. <br> Moreover, in most cases, the driver is not distracted. <br> We can also observe that the points delineated for priors and distractions are fairly close to each other, helping us conclude that there is a large set of drivers involved in accidents who are neither distracted nor were they previously involved in an accident.

#### Bar Graphs

```python
import numpy as np
x=df.loc[(df['ins_premium'] <= 800) & (df['ins_premium'] > 700),["total"]].mean()
y=df.loc[(df['ins_premium'] <= 1000) & (df['ins_premium'] > 800),["total"]].mean()
z=df.loc[(df['ins_premium'] <= 1200) & (df['ins_premium'] > 1000),["total"]].mean()
w=df.loc[(df['ins_premium'] > 1200),["total"]].mean()
a=np.array([x[0],y[0],w[0],z[0]])
b=np.array([">800","800-1000","1000-1200",">1200"])
plt.xlabel("Premium paid")
plt.ylabel("Average number of accidents")
plt.bar(b,a,color='lemonchiffon',width=0.5)
```

**Inference:**<br> The above graph describes the amount of car insurance premium paid. <br> The x axis describes the range in which the premium value lies. The y axis shows the average number of total accidents that have occurred for that amount of premium paid. <br> From the graph, we can deduce that the maximum number of drivers involved in accidents are paying a premium ranging from 800 to 1000 USD.

#### Pie Charts

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])  #[left,bottom,width,height]
axes1.pie(a,labels=b,autopct="%0.2f%%",colors=["mediumspringgreen","mediumaquamarine","aquamarine","turquoise"])
axes1.legend()
```

**Inference:**<br> This pie chart represents the data represented in the bar graph in a visual manner. <br> It is seen that more than a fourth of the drivers pay insurance amounts in the range of 800 to 1000 USD.

#### Histogram

```python
plt.figure().set_figwidth(15)
plt.hist(df.total, bins=20, color='cornflowerblue',edgecolor="white")
plt.xlabel("Total")
plt.ylabel("Count")
t=np.arange(5.9,23.9,0.9)
plt.xticks(t)
```

**Inference:**<br>This histogram chart displays the variation in the total percentage of accidents that have occured. The count axis denotes how many states have total accident percentage lying in the range represented on the x axis<br>We can conclude that the maximum number of states have a total accident percentage lying between 17.6 and 18.5 where as one state has accident percentage as low as between 5.9 and 6.8. A total of 3 states have accident percentage less than 10.4.  

### Using Seaborn to perform visualisation

#### Distplot

```python
sns.distplot(df['total'],color="lightsteelblue")
```

**Inferences:**<br>The given distplot describes the univariate distribution of the values of the total accidents in each American State over its density plot. The distribution attains peak value at a total of 14 accidents(approx) with a value of 0.08.

#### Jointplot

```python
sns.jointplot(x='total',y='alcohol',data=df)
```

**Inference:**<br> A joint plot is used to give a univariate and bivariate analysis of the data at the same time. The bar graphs along the axes describe each of the parameters, while the scatter plot establishes a relationship between the two parameters passed to the plotting function. <br> A moderate positive correlation exists between 'total' and 'alcohol,' with both variables showing increased values together. The right-skewed alcohol histogram indicates concentration on the left side, with visible outliers in both the scatterplot and histograms. The total histogram is not skewed, as values are mostly centrally concentrated.

#### Relplot

```python
sns.relplot(x="not_distracted", y="alcohol", data=df, hue="no_previous",kind="scatter")
```

**Inferences:** The gkven realtional plot analyses the relationship between the drivers from a particular state involved in collisions who were driving under the influence and were not distracted. The observations are futher grouped on the basis of the fact that if the drivers involved in the accidents from a state have any priors.<br>The given plot diaplays a positive relationship between the two parameters. Outliers can be seen as the datapoints which lie away from the majority set.

#### Boxplot

```python
sns.boxplot(data=df[["alcohol", "speeding"]], orient="h")
```

**Inference:**<br>The given box plots are plotted for the alcohol and speeding columns respectively.<br> For both coulmns, majority of the values lie around the 4 to  6 percent region, though speeding has a wider range. Alcohol observations show outliers which are not present in the data for the speeding column.

#### Heatmap

```python
corr = df.corr()
plt.figure(figsize=(7, 5))
sns.heatmap(corr, annot=True, cmap='viridis', linewidths=.5)
```

**Inference:**<br> The above heatmap is plotted for the values of the correlation between each of the columns of the table. <br> A color towards the yellow shade implies a positive and higher correlation value between the variables, whereas a darker, blue-colored cell shows a near-zero or negative correlation value. Larger values of correlation show a dependency between the two variables. A smaller correlation coefficient between two variables establishes that there is no relationship between the two and that the occurrence is merely by chance. <br> This heat map shows that no_previous and total have a strong interdependence, whereas the insurance premium is completely independent of the total accident percentage, the number of accidents caused due to consumption of alcohol, the driver not being distracted, and having no priors. The insurance premium and the losses incurred may be slightly co-dependent.
